### **for loops with bulit-in functions**
- enumerate()
- zip()
- sorted()
- reversed()


image.png

```python
# reversed range
for x in reversed(range(0,5)):
    print(f'for-loop: Number = {x}')
```

```python
# sorted
actor_list = ['Thomas Holland', 
              'Robert Downey', 
              'Benedict Cumberbatch']

for name in actor_list:
    print("original:", name)

print("")

for name in sorted(actor_list):
    print("sorted:", name)
```

```python
# reversed
actor_list = ['Thomas Holland', 
              'Robert Downey', 
              'Benedict Cumberbatch']

for name in actor_list:
    print("original:", name)

print("")

for name in reversed(actor_list):
    print("reversed:", name)
```

```python
# enumerate
actor_list = ['Thomas Holland', 
              'Robert Downey', 
              'Benedict Cumberbatch']

for idx, name in enumerate(actor_list):
    print(idx, name)
```

```python
# zip
actor_list = ['Thomas Holland', 
              'Robert Downey', 
              'Benedict Cumberbatch']
hero_list = ['Spiderman', 
             'Iron Man',
             'Doctor Stange']

for actor, hero in zip(actor_list, hero_list):
    print(f"{hero} is {actor}")
```

### **for loops with lists and nested lists**

```python
# for loop with list
num_list = [0, 1, 2]

for x in num_list:
    print(f'for-loop: Number = {x}')
```

```python
# Nested for
# Processing sublist

world_univ = [['HKUST', 'HKU', 'CUHK'],
              ['NUS', 'NTU', 'SMU'],
              ['MIT', 'USC', 'NYU']]

for sublist in world_univ:
    for u_name in sublist:
        num_letter = len(u_name)
        print("The University: " + u_name + 
              " is " + str(num_letter) + 
              " characters long.")
```

### **break, and pass**

```python
# for, if, break

for letter in 'HKUST':
    print('Current letter :', letter)
    if letter == 'K':        
        break
```

```python
# for, if, break
# .lower()

for letter in 'HKUST'.lower():
    print('Current letter :', letter)
    if letter == 'k':        
        break

for letter in 'HKUST'.lower():
    print('Current letter :', letter)
    if letter.upper() == 'K':
        break
```

```python
# for, pass

for letter in 'HKUST'.lower():
    pass

print('Current letter :', letter)
```

### **continue**

continue is used to skip the current block, and return to the "for" or "while" statement.

```python
# for, continue (WITHOUT continue)
for i in range(5):
    if i > 2:
        print("Ignored",i)
        # continue
        # the following statement will be 
        # skipped when i > 2
        print("Processed",i)
```

```python
# for, continue (WITH continue)
for i in range(5):
    if i > 2:
        print("Ignored",i)
        continue
        # the following statement will be 
        # skipped when i > 2
        print("Processed",i)
```

### **else**

The else block just after for/while is executed only when the loop is NOT terminated by a break statement.

```python
# Example 1: for and else (without break)

for i in range(1, 4): 
    print(i) 
else:  # Executed because no break in for 
    print("No Break") 
```

```python
# Example 2: for and else (with break)

for i in range(1, 4): 
    print(i) 
    break
else: # Not executed as there is a break 
    print("No Break")
```

## **while loops**

image.png

```python
# Test while

i = 1
while i < 3:
    print(i ** 2)
    i = i+1
print('Bye')
```

```python
# Endless loop with break

import time 

i = 1
while True:
    print(f"Hello x{i}")
    time.sleep(1)
    i += 1
    if i == 5:
        break
print("End of an endless loop")
```

```python
# while else (WITHOUT break)

print('WITHOUT break')
n = 5
while n > 0:
    n -= 1
    print(n)
else:
    print('Loop (WITHOUT break): done.')
```

```python
# while else (WITH break)

print('WITH break')
n = 5
while n > 0:
    n -= 1
    print(n)
    break
else:
    print('Loop (WITH break): done.')
```
